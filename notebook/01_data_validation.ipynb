```python
import pandas as pd
```

```python
es_file = '../data/raw/ES.csv'
nq_file = '../data/raw/NQ.csv'
ym_file = '../data/raw/YM.csv'
emd_file = '../data/raw/EMD.csv'
```

# ES (E-Mini S&P 500)

```python
es = pd.read_csv(es_file)

if 'Date' in es.columns:
    es['Date'] = pd.to_datetime(es['Date'])
    es = es.set_index('Date')
```

### Shape

```python
es.shape
```

### NA counts

```python
es.isnull().sum()
```

### Index range

```python
print(es.index.date.min(), " -> ", es.index.date.max())
```
