```python
def trade_signal_test():
    """
    Test trade signal on a mock up DataFrame

    Steps
    -----
    - Load configurations
    - Create a mock up DF
    - Test entry and exit signals on that DF
    
    Returns
    -------

    """

    import yaml 
    import os
    from git import Repo
    repo_root = Repo('.', search_parent_directories=True).working_tree_dir
    os.chdir(repo_root)
    print(repo_root)


    yaml_path = "configs/base.yaml"

    with open(yaml_path, "r") as f:
        config = yaml.safe_load(f)

        
    import pandas as pd
    from src.strategy.signals.composite_rsi_exit import should_exit
    from src.strategy.signals.hurst_filter import allow
    from src.strategy.signals.rsi2 import long_entry 

    df = pd.DataFrame({
        "composite_rsi": [0.48, 0.4, 0.52, 0.38, 0.39],
        "hurst": [0.35, 0.56, 0.57, 0.60, 0.45],
        "rsi": [8, 11, 12, 7, 56]
        })

    params_1 = config["entry_thresholds"]
    params_2 = config["exits"]

    print(df) 
    
    in_trade = False
    
    for i in range(5):
        if(allow(df, i, params_1) and long_entry(df, i, params_1)) and in_trade==False:
            in_trade = True
            print("Long entry at time", i)
        if(should_exit(df, i, params_2)): 
            in_trade = False
            print("Exit at time", i)


trade_signal_test()
```
