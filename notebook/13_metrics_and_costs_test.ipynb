```python
import os
import sys
import pandas as pd
import numpy as np
import yaml

sys.path.append(os.path.abspath(".."))

from src.core.metrics import (
    sharpe_ratio
    sortino_ratio
    max_drawdown
    win_rate
    equity_curve_smoothness
)
from src.core.engine import run
from src.strategy.strategy import Strategy
```

```python
with open("../configs/base.yaml", "r") as f:
    cfg = yaml.safe_load(f)

my_strategy = Strategy(cfg)
```

```python
EMD_path = os.path.join(os.path.abspath(".."), "data", "clean", "EMD_clean.csv")
df_EMD = pd.read_csv(EMD_path)

trades = run(df_EMD, my_strategy)
trades_df = pd.DataFrame(trades)
```

```python
returns = trades_df["net_profit"]
equity = (1 + returns).cumprod()

print("Sharpe:", sharpe_ratio(returns))
print("Sortino:", sortino_ratio(returns))
print("Win rate:", win_rate(returns))
print("Max DD:", max_drawdown(equity))
print("Smoothness:", equity_curve_smoothness(equity))
```
